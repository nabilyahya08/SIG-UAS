```python
# Membaca file geojson dan disimpan ke variabel gdf
gdf = gpd.read_file('RBI25K_ADMINISTRASI_AR.geojson')
# Menampilkan lima baris pertama dari variabel gdf
gdf.head()
```

```python
# Memilih kolom 'NAMOBJ', 'FCODE', 'REMARK', 'METADATA', dan 'geometry' dan menggantikan nilai NaN menjadi 0
gdf = gdf[['NAMOBJ', 'FCODE', 'REMARK', 'METADATA', 'geometry']].fillna(0)
```

```python
# Menginstall modul folium, matplotlib dan mapclassify
!pip install folium matplotlib mapclassify
```

```python
# Mengimport modul folium
import folium

# Membuat map menggunakan folium dan disimpan ke variabel m
m = gdf.explore(column="METADATA", name="Metadata")
# Menambahkan layer control ke objek map dengan variabel m
folium.LayerControl().add_to(m)
# Menampilkan map
m
```
